```python
x_train = train_datagen.flow_from_directory('/home/wsuser/work/dataset/Veg-dataset/Veg-dataset/train_set', 
                                            target_size = (128,128), 
                                            batch_size = 16, 
                                            class_mode = 'categorical')
```

```python
x_test = test_datagen.flow_from_directory('/home/wsuser/work/dataset/Veg-dataset/Veg-dataset/test_set', 
                                          target_size = (128,128), 
                                          batch_size = 16, 
                                          class_mode = 'categorical')
```

```python
from keras.models import Sequential
from tensorflow.keras.models import Sequential
from keras.layers import Dense
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
import warnings
warnings.filterwarnings('ignore')
```

```python
from tensorflow.keras.models import Sequential
model = Sequential()
```

```python

model.add(Convolution2D(32,(3,3), input_shape = (128,128,3), activation = 'relu'))
```

```python
model.add(MaxPooling2D(pool_size = (2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units = 300, activation ='relu'))
```

```python
model.add(Dense(units = 150, activation ='relu'))
```

```python
model.add(Dense(units = 75, activation ='relu'))
```

```python
model.add(Dense(units= 9,activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy',optimizer = "adam",metrics = ["accuracy"])
```

```python
model.fit_generator(x_train,steps_per_epoch = 89,
                    epochs = 5,
                    validation_data = x_test,
                    validation_steps = 27)
```

```python
model.save('vegetable.h5')
```
